```python
import requests
import json
from azure.eventhub import EventHubProducerClient, EventData

# Step 1: Get weather data from OpenWeatherMap API
weather_api_url = "http://api.openweathermap.org/data/2.5/forecast?lat=44.34&lon=10.99&appid=bc6fb4efb13f6c7a51c0a2964c9f20c9"
response = requests.get(weather_api_url)

if response.status_code == 200:
    weather_data = response.json()
else:
    raise Exception(f"API call failed: {response.status_code} - {response.text}")

# Step 2: Define Azure Event Hub connection details
connection_str = "Endpoint=sb://esehsgwa7e7w6ml2w9p0og.servicebus.windows.net/;SharedAccessKeyName=key_139079e7-5062-40c6-8f77-7dcf5f4e0d28;SharedAccessKey=ArOeuw8ZCsrrK9wZjNVAJS49oOf+MBW6u+AEhDU04/4=;EntityPath=es_973d8b9a-ecd8-4da4-b529-f99e90557dd7"

# Step 3: Send the data to Event Hub
producer = EventHubProducerClient.from_connection_string(conn_str=connection_str)

# Convert weather data to JSON string and create EventData
event_data_batch = producer.create_batch()
event_data_batch.add(EventData(json.dumps(weather_data)))

# Send the event
producer.send_batch(event_data_batch)
producer.close()

print("Weather data successfully sent to Event Hub.")
```
